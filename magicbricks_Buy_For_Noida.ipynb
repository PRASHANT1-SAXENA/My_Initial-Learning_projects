```python
import pandas as pd
import numpy as np 
from bs4 import BeautifulSoup
import requests
```

```python
# last page count
page_start=15
group_range=(page_start*30)-30
page_stop=35
itera=1
    
for pg in range(page_start,page_stop):
    url=f"https://www.magicbricks.com/mbsrp/propertySearch.html?editSearch=Y&category=S&propertyType=10000,10006,10012,10005&city=6403&page={page_start}&groupstart={group_range}&offset=0&maxOffset=160&sortBy=premiumRecent&pType=10000,10006,10012,10005&areaUnit=12850&isNRI=N&multiLang=en"
    r=requests.get(url)
    data=r.json()
    a=[]
    for result in data["resultList"]:
        a.append(result["id"])
    t1=a[0]
    
    url_1=f"https://www.magicbricks.com/mbsrp/propertySearch.html?editSearch=Y&category=S&propertyType=10000,10006,10012,10005&city=6403&page={page_start+1}&groupstart={group_range+30}&offset=0&maxOffset=160&sortBy=premiumRecent&pType=10000,10006,10012,10005&areaUnit=12850&isNRI=N&multiLang=en"
  
    r=requests.get(url_1)
    data=r.json()
    b=[]
    for result in data["resultList"]:
        b.append(result["id"])
    t2=b[0]
    if t1==t2:
        print("got it")
        print(t1,t2)
        print(page_start)
        page_end=page_start
        print("page_end: ",page_end)
        break
    else:
        page_start=page_start+1
        group_range=group_range+30
        print(itera,"page",page_start, "Id-",t1)
        itera=itera+1
     
```

```python
mgb = {"unique_id":[],"City":[],"PlotArea":[],"PlotUnit":[],"Rent_price":[],"Dimension":[]
       ,"No_of_open_side":[],"Boundary Wall":[],"Type of Ownership":[],"Overlooking":[]
       ,"Transaction Type":[],"Landmarks":[],"Width of road facing":[],"Pre Leased Property":[]
       ,"Booking Amount":[],"Floors Allowed For Construction":[],"Additional Features":[]
       ,"Facing":[],"Approved By":[],"Any Construction done":[],"ProjectName":[],"Is_Rera":[]
       ,"reraId":[],"propType":[],"comZone":[],"PrimeLoc":[],"tenantsPreference":[]
       ,"latitude":[],"longitude":[],"Property_Name":[],"Property_Address":[],"url":[]}
page_start=1
group_range=(page_start*30)-30
page_stop=2

for pg in range(page_start,page_stop):
    url=f"https://www.magicbricks.com/mbsrp/propertySearch.html?editSearch=Y&category=S&propertyType=10000,10006,10012,10005&city=6403&page={page_start}&groupstart={group_range}&offset=0&maxOffset=160&sortBy=premiumRecent&pType=10000,10006,10012,10005&areaUnit=12850&isNRI=N&multiLang=en"
 
    r=requests.get(url)
    data=r.json()
    for result in data["resultList"]:
#new Attribute  


        try:
            mgb["unique_id"].append(result["id"])
        except:
            mgb["unique_id"].append(np.nan)
         
        try:
            corr=result["ltcoordGeo"].split(",")
            mgb["latitude"].append(corr[0])
            mgb["longitude"].append(corr[1])
        except:
            mgb["latitude"].append(np.nan)
            mgb["longitude"].append(np.nan)
            
        try:
            mgb["Rent_price"].append(result["price"])
        except:
            mgb["Rent_price"].append(np.nan) 
            
        try:
            mgb["City"].append(result["ctName"])
        except:
            mgb["City"].append(np.nan)
            
        
        try:
            mgb["Property_Name"].append(result["propertyTitle"])
        except:
             mgb["Property_Name"].append(np.nan)
                
        try:
            mgb["ProjectName"].append(result["prjname"])
        except:
             mgb["ProjectName"].append(np.nan)
        
        try:
            mgb["Is_Rera"].append(result["isRera"])
        except:
             mgb["Is_Rera"].append(np.nan)
        try:
            mgb["reraId"].append(result["reraId"])
        except:
             mgb["reraId"].append(np.nan)        
        
        
        try:
            mgb["propType"].append(result["propTypeD"])
        except:
             mgb["propType"].append(np.nan)        
         
        try:
            mgb["comZone"].append(result["comZone"])
        except:
             mgb["comZone"].append(np.nan)
        
        try:
            mgb["PrimeLoc"].append(result["isPrimeLocProp"])
        except:
             mgb["PrimeLoc"].append(np.nan)
          
        try:
            mgb["tenantsPreference"].append(result["tenantsPreference"])
        except:
             mgb["tenantsPreference"].append(np.nan)
        
        
        try:
            fetch_url = result["url"]
            url_link = f"https://www.magicbricks.com/propertyDetails/{fetch_url}"
            mgb["url"].append(url_link)
        except:
             mgb["url"].append(np.nan)

        
        r = requests.get(url_link).text
        soup = BeautifulSoup(r, "lxml")
        try:
            title=soup.find_all("div", class_="mb-ldp__more-dtl__list--label")
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div", class_="mb-ldp__more-dtl__list--label")[i].text
                if title=="Address":
                    x.append("yes")
                    address = soup.find_all("div", class_="mb-ldp__more-dtl__list--value")[i].text
                    mgb["Property_Address"].append(address)
                    break
                else:
                    continue
            if "yes"  not in x:
                mgb["Property_Address"].append(address)
            
                                   
        except:
            mgb["Property_Address"].append(np.nan)

#new Attribute----plot area            
                
                
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )[i].text
                if title=="Plot Area":
                    plot_area=soup.find_all("div",class_="mb-ldp__dtls__body__list")[0].contents[0]
                    mgb["PlotArea"].append(plot_area)
                    plot_unit=soup.find_all("div",class_="mb-ldp__dtls__body__list")[0].contents[1].contents[0].string
                    mgb["PlotUnit"].append(plot_unit)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["PlotArea"].append(plot_area)
                mgb["PlotUnit"].append(np.nan)
                
                        
        except:
            mgb["PlotAddress"].append(np.nan)
            mgb["PlotUnit"].append(np.nan)
         
#new Attribute----dimension

        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )[i].text
                if title=="Dimensions(L X B)":
                    x.append("yes")
                    dim=soup.find_all("div",class_="mb-ldp__dtls__body__list--value")[i].text
                    mgb["Dimension"].append(dim)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Dimension"].append(np.nan) 
                        
        except:
            mgb["Dimension"].append(np.nan)             

#new Attribute----   
        
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )[i].text
                if title=="No of Open Sides":
                    x="yes"
                    open_side=soup.find_all("div",class_="mb-ldp__dtls__body__list--value")[i].text
                    mgb["No_of_open_side"].append(open_side)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["No_of_open_side"].append(np.nan)
                 
                        
        except:
            mgb["No_of_open_side"].append(np.nan)
            
    
#new Attribute----
        
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )[i].text
                if title=="Boundary Wall":
                    x="yes"
                    Boundary=soup.find_all("div",class_="mb-ldp__dtls__body__list--value")[i].text
                    mgb["Boundary Wall"].append(Boundary)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Boundary Wall"].append(np.nan)
                 
                        
        except:
            mgb["Boundary Wall"].append(np.nan)

#new Attribute----
        
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )[i].text
                if title=="Type of Ownership":
                    x="yes"
                    ownership=soup.find_all("div",class_="mb-ldp__dtls__body__list--value")[i].text
                    mgb["Type of Ownership"].append(ownership)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Type of Ownership"].append(np.nan)
                 
                        
        except:
            mgb["Type of Ownership"].append(np.nan)

#new Attribute--
        
    
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )[i].text
                if title=="Overlooking":
                    x="yes"
                    overlooking=soup.find_all("div",class_="mb-ldp__dtls__body__list--value")[i].text
                    mgb["Overlooking"].append(overlooking)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Overlooking"].append(np.nan)
                 
                        
        except:
            mgb["Overlooking"].append(np.nan)

#new Attribute--
        
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )[i].text
                if title=="Transaction Type":
                    x="yes"
                    Transaction=soup.find_all("div",class_="mb-ldp__dtls__body__list--value")[i].text
                    mgb["Transaction Type"].append(Transaction)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Transaction Type"].append(np.nan)
                 
                        
        except:
            mgb["Transaction Type"].append(np.nan)
            
#new Attribute--

        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )[i].text
                if title=="Any Construction done":
                    x="yes"
                    Any_Construction_done=soup.find_all("div",class_="mb-ldp__dtls__body__list--value")[i].text
                    mgb["Any Construction done"].append(Any_Construction_done)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Any Construction done"].append(np.nan)
                 
                        
        except:
            mgb["Transaction Type"].append(np.nan)
            

#new Attribute--
        
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__more-dtl__list--label" )[i].text
                if title=="Landmarks":
                    x="yes"
                    Landmark=soup.find_all("div",class_="mb-ldp__more-dtl__list--value")[i].text
                    mgb["Landmarks"].append(Landmark)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Landmarks"].append(np.nan)
                 
                        
        except:
            mgb["Landmarks"].append(np.nan)

#new Attribute--
        
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__more-dtl__list--label" )[i].text
                if title=="Width of road facing":
                    x="yes"
                    width_road=soup.find_all("div",class_="mb-ldp__more-dtl__list--value")[i].text
                    mgb["Width of road facing"].append(width_road)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Width of road facing"].append(np.nan)
                 
                        
        except:
            mgb["Width of road facing"].append(np.nan)
            
#new Attribute--

        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__more-dtl__list--label" )[i].text
                if title=="Pre Leased Property":
                    x="yes"
                    pre_leased =soup.find_all("div",class_="mb-ldp__more-dtl__list--value")[i].text
                    mgb["Pre Leased Property"].append(pre_leased)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Pre Leased Property"].append(np.nan)
                 
                        
        except:
            mgb["Pre Leased Property"].append(np.nan)

#new Attribute--
        
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__more-dtl__list--label" )[i].text
                if title=="Booking Amount":
                    x="yes"
                    Booking =soup.find_all("div",class_="mb-ldp__more-dtl__list--value")[i].text
                    mgb["Booking Amount"].append(Booking)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Booking Amount"].append(np.nan)
                 
                        
        except:
            mgb["Booking Amount"].append(np.nan)
            
#new Attribute--
        
        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__more-dtl__list--label" )[i].text
                if title=="Floors Allowed For Construction":
                    x="yes"
                    floor_for_construction =soup.find_all("div",class_="mb-ldp__more-dtl__list--value")[i].text
                    mgb["Floors Allowed For Construction"].append(floor_for_construction)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Floors Allowed For Construction"].append(np.nan)
                 
                        
        except:
            mgb["Floors Allowed For Construction"].append(np.nan)
            
#new Attribute--

        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__more-dtl__list--label" )[i].text
                if title=="Additional Features":
                    x="yes"
                    Additional_Features =soup.find_all("div",class_="mb-ldp__more-dtl__list--value")[i].text
                    mgb["Additional Features"].append(Additional_Features)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Additional Features"].append(np.nan)
                 
                        
        except:
            mgb["Additional Features"].append(np.nan) 
                      

#new Attribute--


        try:
            title=soup.find_all("div",class_="mb-ldp__dtls__body__list--label" )
            for i in range(len(title)):
                x=[]
                title=soup.find_all("div",class_="mb-ldp__more-dtl__list--label" )[i].text
                if title=="Facing":
                    x="yes"
                    Facing =soup.find_all("div",class_="mb-ldp__more-dtl__list--value")[i].text
                    mgb["Facing"].append(Facing)
                    break
                    
                else:
                    continue
            if "yes" not in x:
                mgb["Facing"].append(np.nan)
                 
                        
        except:
            mgb["Facing"].append(np.nan)
            

#new Attribute--

        try:
            title=soup.find_all("li",attrs={"class":"mb-ldp__dtls__body__summary--item","data-icon": "approved"})[0]
            mgb["Approved By"].append(title.find("span").text)
                                 
        except:
            mgb["Approved By"].append(np.nan)


    
    
        
    for i,j in mgb.items():
        print(i,"=", len(j))
    #print("Page complete -- ",page_start)
    page_start=page_start+1    
    group_range=group_range+30
#mgb_f=pd.DataFrame(mgb)
print("All done")
```
