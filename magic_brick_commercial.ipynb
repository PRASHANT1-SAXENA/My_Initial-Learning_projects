```python
import pandas as pd
import numpy as np
import requests 
from bs4 import BeautifulSoup 
```

```python

mgb = {"commercial_type":[],"url":[],"unique_id":[],"City":[],"Property_Name":[],"Property_Address":[],
       "Commercial_Complex":[],
       "Rent_price":[],"Floor":[],"Units_on_Floor":[],"Pantry":[],"Washroom":[],"Overlooking":[],
       "CarpetArea":[],"CarpetUnit":[],"CoveredArea":[],"Security_Deposit":[],"Maintenance_charges":[],
       "CoveredUnit":[],"latitude":[],"longitude":[],"Facilities":[],"Lifts":[],"Amenities":[],
       "Post_date":[],"Property_Name":[],"Property_Address":[]}

# for office space 
url_1=f"https://www.magicbricks.com/mbsrp/propertySearch.html?editSearch=Y&\
category=R&propertyType=10007,10018&city=6403&page=1&groupstart=28&offset=0\
&maxOffset=62&sortBy=premiumRecent&postedSince=-1&pType=10007,10018&isNRI=N&multiLang=en"

# for shop/showroom
url_2=f"https://www.magicbricks.com/mbsrp/propertySearch.html?editSearch=Y&\
category=R&propertyType=10008,10009&city=6403&page=1&groupstart=28&offset=0\
&maxOffset=16&sortBy=premiumRecent&postedSince=-1&pType=10008,10009&isNRI=N&multiLang=en"

commerecial_types= "office space"
#commerecial_types= "shop/showroom"

r=requests.get(url_1)
data = r.json()
for result in data["resultList"]:
    try:
        mgb["unique_id"].append(result["id"])
        mgb["commercial_type"].append(commerecial_types)
    except:
        mgb["unique_id"].append(np.nan)
        mgb["commercial_type"].append(commerecial_types)
        
        
    try:
        corr=result["ltcoordGeo"].split(",")
        mgb["latitude"].append(corr[0])
        mgb["longitude"].append(corr[1])
    except:
        mgb["latitude"].append(np.nan)
        mgb["longitude"].append(np.nan)
    
    try:
        mgb["Rent_price"].append(result["price"])
    except:
        mgb["Rent_price"].append(np.nan) 
        
   
    try:
        mgb["CarpetArea"].append(result["carpetArea"])
        mgb["CarpetUnit"].append("Sq-ft")
    except:
        mgb["CarpetArea"].append(np.nan)
        mgb["CarpetUnit"].append(np.nan)
            
    try:
        mgb["CoveredArea"].append(result["coveredArea"])
        mgb["CoveredUnit"].append("Sq-ft")
    except:
        mgb["CoveredArea"].append(np.nan)
        mgb["CoveredUnit"].append(np.nan)
    try:
        mgb["City"].append(result["ctName"])
    except:
        mgb["City"].append(np.nan)
    
    try:
        mgb["Property_Name"].append(result["propertyTitle"])
    except:
        mgb["Property_Name"].append(np.nan)
    
            
    try:
        mgb["Post_date"].append(result["postDateT"][0:10])
    except:
        mgb["Post_date"].append(np.nan)
        
    try:
        fetch_url = result["url"]
        url_link = f"https://www.magicbricks.com/propertyDetails/{fetch_url}"
        mgb["url"].append(url_link)
    except:
        mgb["url"].append(np.nan)     
            
    
    r = requests.get(url_link).text
    soup = BeautifulSoup(r, "lxml")
    
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__more-dtl__list--label")[i].text
            if title=="Address":
                address = soup.find_all("div", class_="mb-ldp__more-dtl__list--value")[i].text
                mgb["Property_Address"].append(address)
                break
            else:
                continue
    except:
            mgb["Property_Address"].append(np.nan)
            
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__more-dtl__list--label")[i].text
            if title=="Lifts":
                Lifts = soup.find_all("div", class_="mb-ldp__more-dtl__list--value")[i].text
                mgb["Lifts"].append(Lifts)
                break
            else:
                continue
    except:
            mgb["Lifts"].append(np.nan)      
            
            
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__more-dtl__list--label")[i].text
            if title=="Security Deposit":
                security_Deposit = soup.find_all("div", class_="mb-ldp__more-dtl__list--value")[i].contents[1]
                mgb["Security_Deposit"].append(security_Deposit)
                break
            else:
                continue
    except:
            mgb["Security_Deposit"].append(np.nan)
   

    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__more-dtl__list--label")[i].text
            if title=="Facilities":
                Facilities = soup.find_all("div", class_="mb-ldp__more-dtl__list--value")[i].text
                mgb["Facilities"].append(Facilities)
                break
            else:
                continue
    except:
            mgb["Facilities"].append(np.nan)
    
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__dtls__body__list--label")[i].text
            if title=="Commercial Complex":
                Commercial_Complex = soup.find_all("div", class_="mb-ldp__dtls__body__list--value")[i].text
                mgb["Commercial_Complex"].append(Commercial_Complex)
                break
            else:
                continue
    except:
        mgb["Commercial_Complex"].append(np.nan)
        
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__dtls__body__list--label")[i].text
            if title=="Floor":
                Floor = soup.find_all("div", class_="mb-ldp__dtls__body__list--value")[i].text
                mgb["Floor"].append(Floor)
                break
            else:
                continue
    except:
        mgb["Floor"].append(np.nan)
        
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__dtls__body__list--label")[i].text
            if title=="Units on Floor":
                Units_on_Floor = soup.find_all("div", class_="mb-ldp__dtls__body__list--value")[i].text
                mgb["Units_on_Floor"].append(Units_on_Floor)
                break
            else:
                continue
    except:
        mgb["Units_on_Floor"].append(np.nan)
        
    
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__dtls__body__list--label")[i].text
            if title=="Pantry":
                Pantry = soup.find_all("div", class_="mb-ldp__dtls__body__list--value")[i].text
                mgb["Pantry"].append(Pantry)
                break
            else:
                continue
    except:
        mgb["Pantry"].append(np.nan)
        
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__dtls__body__list--label")[i].text
            if title=="Washroom":
                Washroom = soup.find_all("div", class_="mb-ldp__dtls__body__list--value")[i].text
                mgb["Washroom"].append(Washroom)
                break
            else:
                continue
    except:
        mgb["Washroom"].append(np.nan)
        
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__dtls__body__list--label")[i].text
            if title=="Overlooking":
                Overlooking = soup.find_all("div", class_="mb-ldp__dtls__body__list--value")[i].text
                mgb["Overlooking"].append(Overlooking)
                break
            else:
                continue
    except:
        mgb["Overlooking"].append(np.nan)
        
    try:
        for i in range(0,10):
            title=soup.find_all("div", class_="mb-ldp__dtls__body__list--label")[i].text
            if title=="Maintenance charges":
                Maintenance_charges = soup.find_all("div", class_="mb-ldp__dtls__body__list--value")[i].text
                mgb["Maintenance_charges"].append(Maintenance_charges)
                break
            else:
                continue
    except:
        mgb["Maintenance_charges"].append(np.nan)   
    
   
    try:
        title=soup.find_all("li", class_="mb-ldp__amenities__list--item")
        amme=title[0].text
        for i in range(len(title)):
            if i>0:
                amme= amme+ ", " + title[i].text
        mgb["Amenities"].append(amme)
        
    except:
        mgb["Amenities"].append(np.nan)
        
```

```python
pd.set_option("display.max_columns",100)
df_com=pd.DataFrame(mgb)
print(df_com["url"][0])
print(df_com["url"][3])
df_com
```
