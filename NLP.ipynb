```python
from nltk.tokenize import word_tokenize
from nltk.corpus import gutenberg
from nltk.probability import FreqDist
sample= gutenberg.raw("carroll-alice.txt")
tokenized_word= word_tokenize(sample)
fdist = FreqDist(tokenized_word)
fdist.most_common(20)
#fdist.plot(20)
```

```python
# stop word  are words which occurs most of the documents which convey no meaning which we have to remove
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english")) # set is used to  arragnge in alphbetical order
print(stop_words)
```

```python
# To remove the stopword
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.probability import FreqDist
filterword =[]
sample= gutenberg.raw("carroll-alice.txt")
tokenized_word= word_tokenize(sample)
for i in tokenized_word:
    if i not in stop_words:
        if len(i)>3:
            filterword.append(i)
fdist=FreqDist(filterword)
fdist.most_common(20)
```

```python
# To remove the stopword
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
from nltk.probability import FreqDist
filterword=[i for i in tokenized_word if i not in stop_words and len(i)>3]
fdist=FreqDist(filterword)
fdist.most_common(20)
```

```python
from nltk.corpus import webtext, stopwords
stop_words= set(stopwords.words("english"))
from nltk import bigrams
text_words = [w.lower() for w in webtext.words('pirates.txt') ]
filtered_words= [w for w in text_words if w not in stop_words and len(w)> 3 ]
A= bigrams(filtered_words)
fdist=FreqDist(A)
fdist.most_common(20)
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize,word_tokenize
ps=PorterStemmer()
example= """My name is prashant . I have completed my graduation with math and after that I have
completed the MBA and Right now I am persuing my Postgraduation in Data Science from MIt"""
w_token= word_tokenize(example)
for i in w_token:
    print(ps.stem(i))
```

```python
from nltk.stem import WordNetLemmatizer 
lem=WordNetLemmatizer()
print(lem.lemmatize("cacti"))
print(lem.lemmatize("loins"))
print(lem.lemmatize("loins"))
print(lem.lemmatize("better",pos="a")) # a is used for adjective. and v is used for verb
example= """My name is prashant . I have completed my graduation with math and after that I have
completed the MBA and Right now I am persuing my Postgraduation in Data Science from MIt"""
word=word_tokenize(example)
for i in word:
    print(lem.lemmatize(i))
```

```python
import nltk
from nltk.tokenize import word_tokenize
example= """My name is prashant . I have completed my graduation with math and after that I have
completed the MBA and Right now I am persuing my Postgraduation in Data Science from MIT"""
w_token= word_tokenize(example)
pos_words =nltk.pos_tag(w_token)
pos_words
```

```python
import spacy
# en_core_web_sm is a model 
nlp=spacy.load("en_core_web_sm")
doc=nlp(""" Google Maps, Street View team location across the world, uploading the images
to its website Some have been the scariest, or creeplest, sites ever - including a sun-scorched field in 
the North Carolina.""")
# to print token
for i in doc:
    print(i.text)
# to print token with there respective lemmas
for i in doc:
    print(i.text,i.lemma_)
#to print token , part of speech and tags
for i in doc:
    print(i.text," .... ", i.pos_, " ..... ",i.tag_)
# to get the Entites 
for i in doc.ents:
    print(i.label)
```
